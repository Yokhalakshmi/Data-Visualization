```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("titanic.csv")
```

```python

plt.scatter(data['Age'], data['Fare'])
plt.title('Scatter Plot')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
  
```

```python
sns.lineplot(data['Age'], data['Fare'])
plt.title('Line Chart')
plt.xlabel("Age")
plt.ylabel('Fare')
plt.show()
```

```python
plt.hist(data['Age'],10)
plt.title('Histogram')
plt.xlabel("Age")
plt.ylabel("Range")
plt.show()
```

```python

data.plot.box()
plt.title('Box Plot')
plt.boxplot(data['Age'])
plt.show()
```

```python
data.plot.bar()
  
plt.bar(data['Age'], data['Fare'])
plt.title('Box Plot')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
